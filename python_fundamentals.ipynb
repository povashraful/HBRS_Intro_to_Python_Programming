```python
win = random.randint(1,100)

guess = int(input("Enter a number"))
counter = 1
while guess != win:
  if guess < win:
    print("No- the number is higher - Guess again")
  else:
    print("NO- the number is lower - Guess again")

  guess = int(input("Enter a number"))
  counter +=1

else:
  print("YOU GUESSED it RIGHT- HURRAH")

```
