# Part 2 — Advanced: Introduction to NumPy

NumPy is the foundation of scientific computing in Python. Almost every library you'll meet — SciPy, OpenCV, PyTorch, ROS message helpers, matplotlib — either uses it internally or interoperates with it.

**Why NumPy over plain Python lists?**
- **Speed:** vectorised C implementation, typically 10–100× faster than pure-Python loops.
- **Memory:** compact fixed-type arrays instead of lists of Python objects.
- **Math:** built-in linear algebra, broadcasting, statistics.

## 2.8 Stacking arrays: `vstack` and `hstack`

Sometimes you need to combine several arrays into one — e.g. building a batch of measurements, assembling a matrix from rows/columns, or appending a new sample to a dataset.

- `np.vstack([a, b, ...])` stacks arrays **vertically** (row-wise → new rows added below).
- `np.hstack([a, b, ...])` stacks arrays **horizontally** (column-wise → new columns added to the right).
- `np.concatenate([a, b], axis=...)` is the general version (and what `vstack`/`hstack` call internally).

**Shape rules of thumb**
- `vstack`: arrays must have the same number of **columns**.
- `hstack`: arrays must have the same number of **rows** (for 2-D) or the same length (for 1-D).

**Docs:** [`np.vstack`](https://numpy.org/doc/stable/reference/generated/numpy.vstack.html) · [`np.hstack`](https://numpy.org/doc/stable/reference/generated/numpy.hstack.html) · [`np.concatenate`](https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html)
